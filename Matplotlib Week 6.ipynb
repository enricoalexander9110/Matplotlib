```python
# Belajar Box Plot
```

```python
np.random.seed(2)
data = np.random.normal(loc=100, scale=10, size=200)
data
```

```python
# Simple Box PLot
```

```python
plt.boxplot(data, labels=['Data'])

plt.title('Simple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()

plt.show()
```

```python
# Notched Box Plot
```

```python
plt.boxplot(data, labels=['Data'], notch=True) # tambahkan paramater notch bernilai True

plt.title('Nothced Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
# Pengaturan marker untuk outlier
```

```python
plt.boxplot(data,
            labels = ['Data'],
            showfliers = True,
            flierprops = {'markerfacecolor': 'b', 'marker':'s'})

plt.title('Pengaturan Marker')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
# Horisontal Box Plot
```

```python
plt.boxplot(data, labels=['Data'], vert = False) # (vert = True) = Vertical, (vert = False) = Horisontal)  

plt.title('Horisontal Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
# Multiple Box Plot
```

```python
np.random.seed(10)
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(80, 30, 200)
data3 = np.random.normal(90, 20, 200)
data4 = np.random.normal(70, 25, 200)

data = [data1, data2, data3, data4]
labels = ['Data 1', 'Data 2', 'Data 3', 'Data 4']
```

```python
plt.boxplot(data, labels = labels)

plt.title('Multiple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
#Sample Dataset
```

```python
np.random.seed(2)
data = np.random.normal(loc = 100, scale = 10, size = 200)
data
```

```python
# Simple Violin Plot
```

```python
plt.violinplot(data)

plt.title('Simple Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
# Pengaturan pada Violin Plot
```

```python
plt.violinplot(data,
               showextrema = True, # showextrema = I (berbentuk seperti huruf besar I)
               showmeans = False, # showmeans = nilai rata" (garis tengah)
               showmedians = False, # showmedians = titik yang sama dengan means
               quantiles = [0.25, 0.5, 0.75]) # quantiles = nilai 

plt.title('Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
# Horisontal Violin Plot
```

```python
plt.violinplot(data, vert = False) # False = Horisontal, True = Vertical

plt.title('Horisontal Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
# Kasus Twin Axes 1
```

```python
x = np.linspace(1, 10, 25)
x
```

```python
fig, ax1 = plt.subplots()

ax1.plot(x, np.exp(x), 'bs-', label = 'exp')
ax1.set_xlabel('Sumbu X')
ax1.set_ylabel('exp')

ax2 = ax1.twinx()
ax2.plot(x, np.log(x), 'ro-', label = 'log')
ax2.set_ylabel('log')

fig.suptitle('Contoh Twin Axes')
fig.legend(loc = 'upper left')

plt.show()
```

```python
# Kasus Twin Axes 2 
```

```python
x = np.random.randint(16, 40, size = 30)
x
```

```python
def C2F(celcius = 0):
    return (celcius * 1.8) + 32

def konversi_sumbu(ax1):
    y1, y2 = ax1.get_ylim()
    ax2.set_ylim(C2F(y1), C2F(y2))
    ax2.figure.canvas.draw()
    
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
    
ax1.callbacks.connect('ylim_changed', konversi_sumbu)
ax1.plot(x)
    
ax1.set_xlabel('Hari')
ax1.set_ylabel('Celcius')
ax2.set_ylabel('Fahrenheit')
    
fig.suptitle('Temperatur Udata')

plt.show()
```
