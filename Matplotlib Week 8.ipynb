```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
# Matplotlib 16 | Alur Hidup Plot 
```

```python
data = {'Item A' : 109438,
       'Item B' : 103569,
       'Item C' : 112214,
       'Item D' : 112591,
       'Item E' : 100934,
       'Item F' : 103660,
       'Item G' : 137351,
       'Item H' : 123381,
       'Item I' : 135841,
       'Item J' : 100437}

data
```

```python
items = tuple(data.keys())
items
```

```python
count = tuple(data.values())
count
```

```python
# Simple Plot
```

```python
fig, ax = plt.subplots()

ax.barh(items, count)

plt.show()
```

```python
# Pengaturan Style
```

```python
plt.style.available
```

```python
plt.style.use('bmh')

fig, ax = plt.subplots()
ax.barh(items, count)
plt.show()
```

```python
# Pengaturan Tick Label
```

```python
fig, ax = plt.subplots()
ax.barh(items, count)

labels = ax.get_xticklabels()

# set properties
plt.setp(labels,
        rotation = 45,
        horizontalalignment = 'right')

plt.show()
```

```python
from matplotlib.ticker import FuncFormatter
```

```python
def ribuan(x, pos):
    return f'{int(x/1000)}K'
```

```python
fig, ax = plt.subplots()
ax.barh(items, count)

formatter = FuncFormatter(ribuan)
ax.xaxis.set_major_formatter(formatter)

ax.set(title = 'Contoh Plot',
      xlabel = 'Jumlah',
      ylabel = 'Items')

plt.show()
```

```python
# Pengaturan Label pada Sumbu (axis) dan Judul (title)
```

```python
fig, ax = plt.subplots()
ax.barh(items, count)

ax.set(title= 'Contoh Plot',
      xlabel = 'Jumlah',
      ylabel = 'Items')

plt.show()
```

```python
# Penambahan Garis (vertical / horisontal line) pada Plot
```

```python
fig, ax = plt.subplots()
ax.barh(items, count)

ax.axvline(np.mean(count),
          ls = '-',
          color = 'g')

ax.set(title= 'Contoh Plot',
      xlabel = 'Jumlah',
      ylabel = 'Items')

plt.show()
```

```python
fig.canvas.get_supported_filetypes()
```

```python
fig, ax = plt.subplots()
ax.barh(items, count)

ax.axvline(np.mean(count),
          ls = '--',
          color = 'r')

ax.set(title= 'Contoh Plot',
      xlabel = 'Jumlah',
      ylabel = 'Items')

fig.savefig('sales.png',
           transparent = False,
           dpi = 80)
```
