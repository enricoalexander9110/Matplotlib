```python
plt.plot([2, 5, 7, 11])
plt.ylabel('sumbu y')
plt.show()
```

```python
# g = green , b = blue , r = red | - = garis , o = titik , * = bintang
```

```python
plt.plot([1, 2, 3, 4] , [1, 4, 9, 16], 'g*')
plt.axis([0, 6, 0, 20])
plt.show()
```

```python
# Multi plot dalam satu pemanggilan funsi plot()
```

```python
t = np.arange(0., 5., 0.2)
t
```

```python
plt.plot(t, t, 'r--')
plt.plot(t, t**2, 'bs')
plt.plot(t, t**3, 'g^')
plt.show()
```

```python
# Kita juga dapat melakukan multi plot dalam satu pemanggilan fungsi plot()
```

```python
plt.plot(t, t, 'r--',
         t, t**2, 'bs',
         t, t**3, 'g^')
plt.show()
```

```python
# Plotting dengan keyboard
```

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

data
```

```python
plt.scatter('a', 'b', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```
